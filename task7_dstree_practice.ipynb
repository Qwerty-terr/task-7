```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return np.sin(x)   # истинная зависимость в данных
sample_size = 100      # длина выборки
samples_num = 20       # количество выборок
linspace = np.linspace(0, 7, 1000)  # точки для построения графиков
```

__1. (2 балла)__
1. Сгенерируйте выборку $x$ из одномерного экспоненциального распределения (np.random.exponential) длины sample_size.
1. Создайте вектор целевых переменных $y$ как сумму $f(x)$ и случайного шума, сгенерированного из равномерного распределения на отрезке $[-1, 1]$ (np.random.uniform).
1. Обучите DecisionTreeRegressor с параметрами по умолчанию на полученной выборке и сделайте предсказания для объектов из linspace.
1. Постройте два графика на одном рисунке: $f(x)$ и зависимость, восстановленную решающим деревом.

    Рекомендация: не забудьте, что все методы обучения в sklearn требуют на вход двумерную матрицу объекты-признаки. Сделать такую из одномерного вектора можно добавлением мнимых осей (np.newaxis).

```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np
from matplotlib import pyplot as plt

x = np.random.exponential(scale=1, size=sample_size)
noise = np.random.uniform(low=-1, high=1, size=sample_size)
y = np.sin(x) + noise

regressor = DecisionTreeRegressor()
regressor.fit(x[:, np.newaxis], y)

y_pred = regressor.predict(linspace[:, np.newaxis])

plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='с шумом', color='red', alpha=0.5)
plt.plot(linspace, np.sin(linspace), color='green', linestyle='--', label='sin(x)')
plt.plot(linspace, y_pred, label='Decision Tree Regressor', color='r')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

__3.(0.5 балл):__
Повторите предыдущий пункт, установив максимальную глубину решающего дерева равной 2, а затем равной 4. Таким образом, у вас получится еще два графика.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

def get(sample_size=100):
    x = np.random.exponential(1, sample_size)
    y = np.sin(x) + np.random.uniform(-1, 1, sample_size)
    return x[:, np.newaxis], y

def depth(x, y, depth_value, grid):
    model = DecisionTreeRegressor(max_depth=depth_value)
    model.fit(x, y)
    return model.predict(grid)

samples_num = 10
linspace = np.linspace(0, 7, 1000) [:, np.newaxis]

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))

# Левый
y_preds = np.zeros((samples_num, len(linspace)))
for i in range(samples_num):
    x, y = get()
    y_preds[i] = depth(x, y, 2, linspace)
    ax1.plot(linspace, y_preds[i], color='gray', alpha=0.5)
    
ax1.plot(linspace, np.sin(linspace), color='red', label='f(x)')
ax1.plot(linspace, np.mean(y_preds, 0), color='black', label='Усредненная')
ax1.set_title('Глубина дерева = 2')

# Правый
y_preds = np.zeros((samples_num, len(linspace)))
for i in range(samples_num):
    x, y = get()
    y_preds[i] = depth(x, y, 4, linspace)
    ax2.plot(linspace, y_preds[i], color='gray', alpha=0.5)

ax2.plot(linspace, np.sin(linspace), color='red', label='f(x)')
ax2.plot(linspace, np.mean(y_preds, 0), color='black', label='Усредненная')
ax2.set_title('Глубина дерева = 4')

for ax in [ax1, ax2]:
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.legend()

plt.tight_layout()
plt.show()
```
